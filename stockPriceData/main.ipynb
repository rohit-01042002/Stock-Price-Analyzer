```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from datetime import datetime
```

```python
import os
companies=os.listdir()[1:]
```

```python
companyData=dict()
for company in companies:
    if company.split('.')[-1]=='csv':
        data=pd.read_csv(company)
        data=data.dropna()
        companyData[company]=data
```

```python
def analyse(company):
    file=str.upper(company)+'.csv'
    df=companyData[file]
    df['Date']=pd.to_datetime(df['Date'],format="%Y-%m-%d")
    
```
